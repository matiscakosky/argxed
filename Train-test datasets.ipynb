```python
CATEGORIA_ACTUAL = "nivel_desemp_ciencias_naturales"
PONDERADOR_ACTUAL = "ponderador_ciencias_naturales"
CATEGORIA_NIVEL = "ciencias_naturales"
```

```python
actual_full = full_dataset[['alumne_id', 'sexo', 'indice_socioeconomico', CATEGORIA_ACTUAL,
       PONDERADOR_ACTUAL, 'tiene_notebook', 'tiene_pc',
       'tiene_tablet', 'tiene_celular', 'tiene_smartphone', 'tiene_consola',
       'tiene_smarttv', 'tiene_cable', 'tiene_internet_alumne',
       'repeticion_primaria', 'repeticion_secundaria', 'nivel',
       'ambito', 'gestion', 'icse', 'icse_cat', 'indice_socioeconomico_medio',
       'indice_socioeconomico_medio_cat', 'tecnica', 'tiene_internet_escuela',
       'subvencion', 'dependencia', 'name_provincia', 'departamento_name',
       'tasa_con_nbi']]
```

```python
actual_full = actual_full[((actual_full[CATEGORIA_ACTUAL].notnull()) & (actual_full[CATEGORIA_ACTUAL]!="nc") & (actual_full[PONDERADOR_ACTUAL]!=-1.0))]
actual_full = actual_full.reset_index(drop=True)
```

```python
SEXO_MAP = {"nc": 0, "f":1, "m":2}
SEXO_NAMES = ["Sexo: No contesta", "Sexo: Femenino", "Sexo: Masculino"]
sexo_onehot = np.array([ np.eye(len(SEXO_MAP))[SEXO_MAP[s]] if isinstance(s, str) else np.zeros(len(SEXO_MAP)) for s in actual_full["sexo"] ])
sexo_df = pd.DataFrame(sexo_onehot, columns=SEXO_NAMES)
sexo_df.sample(5)
```

```python
INDICE_SOCIOECONOMICO_MAP = {"nc": 0, "bajo":1, "medio":2, "alto":3}
INDICE_SOCIOECONOMICO_NAMES = ["Indice Socioeconomico: No contesta", "Indice Socioeconomico: Bajo", "Indice Socioeconomico: Medio",
                               "Indice Socioeconomico: Alto"]
indice_socioeconomico_onehot = np.array([ np.eye(len(INDICE_SOCIOECONOMICO_MAP))[INDICE_SOCIOECONOMICO_MAP[s]] if isinstance(s, str) else np.zeros(len(INDICE_SOCIOECONOMICO_MAP)) for s in actual_full["indice_socioeconomico"] ])
indice_socioeconomico_df = pd.DataFrame(indice_socioeconomico_onehot, columns=INDICE_SOCIOECONOMICO_NAMES)
indice_socioeconomico_df.sample(5)
```

```python
TEC_COLUMNS = ['tiene_notebook', 'tiene_pc', 'tiene_tablet', 'tiene_celular', 'tiene_smartphone', 
               'tiene_consola','tiene_smarttv', 'tiene_cable', 'tiene_internet_alumne']
```

```python
tec_columns = actual_full[TEC_COLUMNS]
for t in TEC_COLUMNS:
    col_name = "Alumne "+" ".join(t.split("_")[:2])
    tec_columns[col_name+": "+"No contesta"] = tec_columns[t].map(lambda x: 1 if x == "nc" else 0)
    tec_columns[col_name+": "+"No"] = tec_columns[t].map(lambda x: 1 if x == "no" else 0)
    tec_columns[col_name+": "+"Sí"] = tec_columns[t].map(lambda x: 1 if x == "sí" else 0)
tec_columns = tec_columns[[c for c in tec_columns.columns if c not in TEC_COLUMNS]]
tec_columns.sample(5)
```

```python
REPETICIONES_COLUMNS = ["repeticion_primaria", "repeticion_secundaria"]
```

```python
repeticiones_columns = actual_full[REPETICIONES_COLUMNS]
for t in REPETICIONES_COLUMNS:
    col_name = "Alumne "+" ".join(t.split("_")[:2])
    repeticiones_columns[col_name+": "+"No contesta"] = repeticiones_columns[t].map(lambda x: 1 if x == "nc" else 0)
    repeticiones_columns[col_name+": "+"No"] = repeticiones_columns[t].map(lambda x: 1 if x == "no" else 0)
    repeticiones_columns[col_name+": "+"Una vez"] = repeticiones_columns[t].map(lambda x: 1 if x == "una_vez" else 0)
    repeticiones_columns[col_name+": "+"Des veces"] = repeticiones_columns[t].map(lambda x: 1 if x == "una_vez" else 0)
    repeticiones_columns[col_name+": "+"Tres veces o más"] = repeticiones_columns[t].map(lambda x: 1 if x == "una_vez" else 0)
repeticiones_columns = repeticiones_columns[[c for c in repeticiones_columns.columns if c not in REPETICIONES_COLUMNS]]
repeticiones_columns.sample(5)
```

```python
AMBITO_MAP = {"urbano": 0, "rural":1}
AMBITO_NAMES = ["Ámbito: Urbano", "Ámbito: Rural"]
ambito_onehot = np.array([ np.eye(len(AMBITO_MAP))[AMBITO_MAP[s]] if isinstance(s, str) else np.zeros(len(AMBITO_MAP)) for s in actual_full["ambito"] ])
ambito_df = pd.DataFrame(ambito_onehot, columns=AMBITO_NAMES)
ambito_df.sample(5)
```

```python
GESTION_MAP = {"estatal": 0, "privada":1, "social/cooperativa": 2}
GESTION_NAMES = ["Gestión: Estatal", "Gestión: Privada", "Gestión: Social/Cooperativa"]
gestion_onehot = np.array([ np.eye(len(GESTION_MAP))[GESTION_MAP[s]] if isinstance(s, str) else np.zeros(len(GESTION_MAP)) for s in actual_full["gestion"] ])
gestion_df = pd.DataFrame(gestion_onehot, columns=GESTION_NAMES)
gestion_df.sample(5)
```

```python
ICSE_MAP = {"critico": 0, "alto":1, "bajo_y_medio": 2}
ICSE_NAMES = ["ICSE: Crítico", "ICSE: Alto", "ICSE: Bajo y Medio"]
icse_onehot = np.array([ np.eye(len(ICSE_MAP))[ICSE_MAP[s]] if isinstance(s, str) else np.zeros(len(ICSE_MAP)) for s in actual_full["icse_cat"] ])
icse_df = pd.DataFrame(icse_onehot, columns=ICSE_NAMES)
icse_df["ICSE"] = actual_full["icse"]
icse_df["ICSE"] = icse_df["ICSE"].fillna(-1.0)
icse_df["ICSE_null"] = icse_df["ICSE"].map(lambda x: 1 if x==-1.0 else 0)
icse_df.sample(5)
```

```python
np.random.seed(311997)
features_full = features_full.iloc[np.random.permutation(len(features_full))]
test = features_full[:math.floor(len(features_full)*0.1)]
train = features_full[math.floor(len(features_full)*0.1):]
```

```python
def ponderar(dataframe, ponderador):
    no_ponderar = dataframe[dataframe[ponderador]==1.0]
    ponderadas = []
    a_ponderar = dataframe.get_values()[dataframe.get_values()[:,-2]>1.0].tolist()
    for row in a_ponderar:
        multiplicador = math.floor(row[-2])
        proba_excedente = row[-2] - multiplicador
        multiplicador += np.random.binomial(1,proba_excedente)
        ponderadas += [row]*multiplicador
    return pd.concat([no_ponderar, pd.DataFrame(ponderadas, columns=dataframe.columns)], axis=0)
```

```python
ponderar(train[train[PONDERADOR_ACTUAL]>1.0][:5], PONDERADOR_ACTUAL)
```

```python
train=ponderar(train, PONDERADOR_ACTUAL)
test=ponderar(test, PONDERADOR_ACTUAL)
```

```python
train.to_csv(FEATURES_CSVS_PATH+"train_%s.csv" % CATEGORIA_NIVEL, index=False)
test.to_csv(FEATURES_CSVS_PATH+"test_%s.csv" % CATEGORIA_NIVEL, index=False)
```
