```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
full_dataset = pd.read_csv("super_join.csv")
```

```python
full_dataset.columns
```

# Por nivel

```python
plt.figure(dpi=100)
plt.title("Desempeño de lengua segun nivel")
g = sns.countplot(x="nivel_desemp_lengua", hue="nivel", data=full_dataset[(full_dataset["nivel_desemp_lengua"]!="nc")], 
                  order=["avanzado", "satisfactorio", "basico", "por_debajo_del_basico"])
plt.xlabel("Nivel de desempeño en lengua")
plt.ylabel("Cantidad")
leg = g.axes.get_legend()
leg.set_title("Nivel educativo")
LABELS = {"primaria": "Primaria", "secundaria": "Secundaria"}
for t in leg.texts: t.set_text(LABELS[t.get_text()])
plt.xticks(rotation=90)
plt.savefig("plots/desempeño_lengua_nivel.png")
```
